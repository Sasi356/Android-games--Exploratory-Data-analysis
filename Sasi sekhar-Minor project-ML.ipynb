```python
from pandas import*
from numpy import*
from datetime import*
import time as t
print(data['In-app Purchases'].isnull().sum())
data=read_csv('appstore_games (5).csv')
data=data.drop(['URL','ID','Description','Icon URL','Subtitle','Developer','Primary Genre'],axis=1)
data=data.dropna(subset=['User Rating Count'])
data=data.dropna()
data.head(10)
data['Genres'].unique()
data.head()
```

```python
data1=data['Genres'].str.split(",",expand=True)
data1.columns =['q','w','e','r']
for i in ['q','w','e','r']:
    data2=data1[i].str.replace("Games","")
    data1=data1.drop([i],axis=1)
    data1[i]=data2
for i in ['w','e','r','q']:
    data2=data1[i].str.replace("Entertainment","")
    data1=data1.drop([i],axis=1)
    data1[i]=data2
for i in ['w','e','r','q']:
    data2=data1[i].str.replace("Utilities","")
    data1=data1.drop([i],axis=1)
    data1[i]=data2
for i in ['w','e','r','q']:
    data2=data1[i].str.replace("Strategy","")
    data1=data1.drop([i],axis=1)
    data1[i]=data2
for i in ['w','e','r','q']:
    data2=data1[i].str.replace("Role Playing","Role_Playing")
    data1=data1.drop([i],axis=1)
    data1[i]=data2
for i in ['w','e','r','q']:
    data2=data1[i].str.replace("Social Networking","Social_Networking")
    data1=data1.drop([i],axis=1)
    data1[i]=data2
for i in ['w','e','r','q']:
    data2=data1[i].str.replace("Health & Fitness","Health_&_Fitness")
    data1=data1.drop([i],axis=1)
    data1[i]=data2
for i in ['w','e','r','q']:
    data2=data1[i].str.replace("Food & Drink","Food_&_Drink")
    data1=data1.drop([i],axis=1)
    data1[i]=data2
for i in ['w','e','r','q']:
    data2=data1[i].str.replace("Racing","Adventure")
    data1=data1.drop([i],axis=1)
    data1[i]=data2
for i in ['w','e','r','q']:
    data2=data1[i].str.replace("Finance","Family")
    data1=data1.drop([i],axis=1)
    data1[i]=data2
    
data1.head(10)
```

```python
dat=data1['Genres'].str.split(' ',expand=False)
for i in dat.keys():
    if 'float' in str(type(dat[i])):
        continue
    for j in dat[i]:
        if j == '' or j == ' ':
            dat[i].remove(j)
    for j in dat[i]:
        if j == '' or j == ' ':
            dat[i].remove(j)    
dat.tail(50)
```

```python
for i in dat.keys():
    if 'float' in str(type(dat[i])):
        continue
    for j in dat[i]:
        if j== 'Board':
            dat[i].remove(j)
            dat[i].append('Puzzle')
        if j== 'Playing':
            dat[i].remove(j)
            dat[i].append('Puzzle')
        if j== 'Role':
            dat[i].remove(j)
            dat[i].append('Adventure')
        if j== 'Role_Playing':
            dat[i].remove(j)
            dat[i].append('Adventure')
        if j== 'Education':
            dat[i].remove(j)
            dat[i].append('Family')
        if j== 'Card':
            dat[i].remove(j)
            dat[i].append('Family')
        if j== 'Photo_&_Video':
            dat[i].remove(j)
            dat[i].append('Family')
        if j== 'Food_&_Drink':
            dat[i].remove(j)
            dat[i].append('Family')
dat
```
